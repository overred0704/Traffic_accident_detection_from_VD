##generate raw data

```python
def raw():
  !gdown --id '1VTTNmFoXONE5dIsU9UDuvcisc419LEGd' --output file.csv
  ttt = pd.read_csv('file.csv')
  #基本欄位轉換和命名
  #這邊不用動
  ttt.columns=['id','VDID','LinkID','Speed','Occupancy','Volume']
  a = ttt.drop('id',axis=1)
  Road = a.VDID.str.split('-').str.get(1)
  Direction = a.VDID.str.split('-').str.get(2)
  Mile = a.VDID.str.split('-').str.get(3)
  a['Road'] = Road
  a['Direction'] = Direction
  a['Mile'] = Mile
  a = a[['Road','Direction','Mile','Speed','Occupancy','Volume','VDID','LinkID']]
  for i in range(len(a)):
    try:
      a['Mile'][i] = float(a['Mile'][i])
    except:
      pass
  a['Mile'][1686] = float(0.008)
  a['Mile'][1992] = float(0.100)  
  return a
```

##generate table

```python
def null_table():
  lst = []
  for i in range(1440):
    lst.append(i)
  database = pd.DataFrame(index=lst)
  database['Time'] = None

  for i in range(1,8):
    database['b_'+str(i)+'_s'] = None
    database['b_'+str(i)+'_v'] = None
    database['b_'+str(i)+'_o'] = None
    database['f_'+str(i)+'_s'] = None
    database['f_'+str(i)+'_v'] = None
    database['f_'+str(i)+'_o'] = None
    database['avg_'+str(i)+'_s'] = None
    database['avg_'+str(i)+'_v'] = None
    database['avg_'+str(i)+'_o'] = None
    database['dif_'+str(i)+'_s'] = None
    database['dif_'+str(i)+'_v'] = None
    database['dif_'+str(i)+'_o'] = None
  
  database['Crush'] = None
  return database
```

##build dataset

```python
def inference_data_build(accdata, acctime):
  #set the df
  ft = accdata
  a = raw()

  #先假設要抓這個
  filt =  (ft['Time'] == acctime)
  ft[filt]
  vava = ft[filt].index.values
  ind = int(vava)

  #搜尋VD_INDEX
  temp_list = pd.DataFrame(columns=('Road','Direction','Mile','Speed','Occupancy','Volume','VDID','LinkID'))
  for i in range(len(a)):
    if ft['Roadname'][ind] == a['Road'][i] and ft['Direction'][ind] == a['Direction'][i]:
      #print(i)
      temp_list = temp_list.append(a[i:i+1])

  #sort temp list
  temp_list = temp_list.sort_values(by=['Mile'])
  temp_list = temp_list.reset_index(drop=True)
  
  #set the collistion time
  collisiontime = 0
  for j in range(len(temp_list)):
    if temp_list['Mile'][j] <= ft['Mile'][ind] <= temp_list['Mile'][j+1]:
      #print(j)
      collisiontime = j
  
  print(temp_list['VDID'][collisiontime])
  print(temp_list['VDID'][collisiontime+1])
  vdvd = temp_list['VDID'][collisiontime]

  #get none table
  database = null_table()
  print('-----------------writing-----------------')

  #start!
  k = 0
  while k <= 24*60:
    time = acctime[0:4] + str('0000')
    now = dic.index(time)
    p = now + k
  
    database['Time'][k] = dic[p]

    for j in range(1,8):
      path = '04DL/'+ dic[p-5] +'.csv'
      try:
        file = pd.read_csv(path)
      except:
        print(path, 'is not found')
        continue
      else:

        print(dic[p-5])
        p += 1

      
      for i in range(len(file)):
        if file['VDID'][i] == vdvd:
          VD_index = i
          #print('VD_index is :',i)

      #print(file['VDID'][VD_index])
      back_1 = file[file['VDID'] == file['VDID'][VD_index]]
      front_1 = file[file['VDID'] == file['VDID'][VD_index + 1]]

      database['b_' + str(j) + '_s'][k] = float(back_1['Speed'].values)  
      database['b_' + str(j) + '_v'][k] = float(back_1['Volume'].values)
      database['b_' + str(j) + '_o'][k] = float(back_1['Occupancy'].values)

      database['f_' + str(j) + '_s'][k] = float(front_1['Speed'].values)   
      database['f_' + str(j) + '_v'][k] = float(front_1['Volume'].values)
      database['f_' + str(j) + '_o'][k] = float(front_1['Occupancy'].values)

      database['avg_' + str(j) + '_s'][k] = ((float(back_1['Speed'].values)) + (float(front_1['Speed'].values))) / 2
      database['avg_' + str(j) + '_v'][k] = ((float(back_1['Volume'].values)) + (float(front_1['Volume'].values))) / 2
      database['avg_' + str(j) + '_o'][k] = ((float(back_1['Occupancy'].values)) + (float(front_1['Occupancy'].values))) / 2

      database['dif_' + str(j) + '_s'][k] = abs((float(back_1['Speed'].values)) - (float(front_1['Speed'].values)))
      database['dif_' + str(j) + '_v'][k] = abs((float(back_1['Volume'].values)) - (float(front_1['Volume'].values)))
      database['dif_' + str(j) + '_o'][k] = abs((float(back_1['Occupancy'].values)) - (float(front_1['Occupancy'].values)))

    
    if dic[p] == acctime :
      database['Crush'][k] = 1
    else:
      database['Crush'][k] = 0


      
    
    k += 1
    print('--------------------------------')
    
  database = database.drop('Crush', axis=1)
  d = database.dropna(axis=0)
  for cols in d.columns.tolist()[1:]:
    d = d[d[cols] >= 0]

  print('Finish!!!')
  return d
```

```python
def ML_inference(data, model, acctime, duration):
  #下載原本資料分布
  !gdown --id '1Laqz5Z_5FWPOiK42oUkqS4AVnH8zcpLN' --output ori.csv
  ori = pd.read_csv('ori.csv')
  ooo = ori.iloc[:,62:146]

  #set time list
  time = data['Time']
  data = data.drop('Time', axis=1)

  #set the ori time list
  rawtime = acctime[0:4] + str('0000')
  for i,k in enumerate(dic):
    if rawtime == k:
      index = i
  raw_list = []
  for i in range(1440):
    raw_list.append(dic[index])
    index += 1

  #set during time set
  ind = raw_list.index(acctime)
  s = set()
  for i in range(int(duration)):
    s.add(raw_list[ind+i])



  #轉換欄位名稱
  col1 = ooo.columns
  col2 = data.columns 

  for i in range(84):
    c1 = col1[i]
    c2 = col2[i]
    data[c2] = (data[c2] - ooo[c1].min())/\
            (ooo[c1].max() - ooo[c1].min())
    #print(i, c1, c2)
  #data = data.drop('Crush', axis=1)
  
  #轉換欄位名稱
  md = joblib.load('xg_SMOTE_model.joblib')
  cols_when_model_builds = md.get_booster().feature_names 
  data.columns = cols_when_model_builds

  #inference
  y_pred = model.predict(data)
  

  #prepare
  t2 = []
  for i in time:
    q = str(0) + str(i)
    t2.append(q)

  pred = list(y_pred)

  inddd = 0

  #count
  count=0
  vderror = 0

  #print result
  for i in raw_list:
    if i in t2:
      if pred[inddd] == 0:
        print('time:', t2[inddd] , 'Nothing')
        inddd += 1
      else:
        print('time:', t2[inddd] , 'Caution! May be accident occurrence!')
        inddd += 1
        count += 1
        if t2[inddd] in s:
          print('!!!!!!This is real accident case!!!!!!')
          count -= 1
    else:
      vderror += 1
      print('time:', i, 'VD error!!')



  print('-------------------------Summary----------------------------')

  print('VD error count: ', vderror)
  print('VD error rate: ', (vderror/1440)*100 , '%')
  print('False alarm count: ', count)
  print('FAR: ', (count/1440*100), '%')
```

```python
def DL_inference(data, model, acctime, duration):
  #下載原本資料分布
  !gdown --id '1Laqz5Z_5FWPOiK42oUkqS4AVnH8zcpLN' --output ori.csv
  ori = pd.read_csv('ori.csv')
  ooo = ori.iloc[:,62:146]

  #set time list
  time = data['Time']
  data = data.drop('Time', axis=1)

  #set the ori time list
  rawtime = acctime[0:4] + str('0000')
  for i,k in enumerate(dic):
    if rawtime == k:
      index = i
  raw_list = []
  for i in range(1440):
    raw_list.append(dic[index])
    index += 1

  #set during time set
  ind = raw_list.index(acctime)
  s = set()
  for i in range(int(duration)):
    s.add(raw_list[ind+i])


  #轉換欄位名稱
  col1 = ooo.columns
  col2 = data.columns 

  for i in range(84):
    c1 = col1[i]
    c2 = col2[i]
    data[c2] = (data[c2] - ooo[c1].min())/\
            (ooo[c1].max() - ooo[c1].min())
    #print(i, c1, c2)
  #data = data.drop('Crush', axis=1)
  
  #轉換欄位名稱
  md = joblib.load('xg_SMOTE_model.joblib')
  cols_when_model_builds = md.get_booster().feature_names 
  data.columns = cols_when_model_builds

  #convert to 3 dimension
  test_time_x = data
  test_time_x = test_time_x.to_numpy()
  test_time_x = test_time_x .reshape((test_time_x.shape[0], 7 ,12))

  #inference
  y_pred = model.predict_classes(test_time_x)


  #prepare
  t2 = []
  for i in time:
    q = str(0) + str(i)
    t2.append(q)

  pred = list(y_pred)
  inddd = 0

  #count
  count=0
  vderror = 0

  #print result
  for i in raw_list:
    if i in t2:
      if pred[inddd] == 0:
        print('time:', t2[inddd] , 'Nothing')
        inddd += 1
      else:
        print('time:', t2[inddd] , 'Caution! May be accident occurrence!')
        inddd += 1
        count += 1
        if t2[inddd] in s:
          print('!!!!!!This is real accident case!!!!!!')
          count -= 1
    else:
      vderror += 1
      print('time:', i, 'VD error!!')



  print('-------------------------Summary----------------------------')

  print('VD error count: ', vderror)
  print('VD error rate: ', (vderror/1440)*100 , '%')
  print('False alarm count: ', count)
  print('FAR: ', (count/1440*100), '%')
```
