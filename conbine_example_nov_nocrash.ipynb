```python
#確認原始的事故時間
oritime = []
for i in range(376):
  mon = nov['月'][i]
  mon = str(mon)

  day = nov['日'][i]
  if day < 10:
    day = str(0) + str(day)
  else:
    day = str(day)

  hr = nov['時'][i]
  if hr < 10:
    hr = str(0) + str(hr)
  else:
    hr = str(hr)

  min = nov['分'][i]
  if min < 10:
    min = str(0) + str(min)
  else:
    min = str(min)

  tt = mon + day + hr + min
  oritime.append(tt)
```

```python
time = []
for i in range(376):
  mon = nov['月'][i]
  mon = str(mon)

  day = nov['日'][i]
  if day < 10:
    day = str(0) + str(day)
  else:
    day = str(day)

  hr = nov['時'][i]+ 2
  if hr < 10:
    hr = str(0) + str(hr)
  elif hr > 23:
    hr = str('00')

    day = nov['日'][i]+1
    if day < 10:
      day = str(0) + str(day)
    else:
      day = str(day)

  else:
    hr = str(hr)

  min = nov['分'][i]
  if min < 10:
    min = str(0) + str(min)
  else:
    min = str(min)

  tt = mon + day + hr + min
  time.append(tt)
print(time)
```

```python
#確認是否重複:
rep = []
for i in range(len(oritime)):
  for j in range(len(time)):
    if time[j] == oritime[i]:
      print('重疊的時間點:',time[j],', 原始事故時間:', oritime[i])
      print('---------------------')
      rep.append(time[j])
```

```python
col_name=nov.columns.tolist() 
col_name.insert(5,'time') 
nov=nov.reindex(columns=col_name)   
nov['time']=time
```

```python
nov = nov.drop(['年','月','日','時','分','事件發生','事件排除'],axis=1)
nov = nov[['id','time','國道名稱','方向','里程','處理分鐘','事故類型','死亡','受傷']]
nov['國道名稱'] = nov['國道名稱'].str.replace('國','N')
nov = nov.replace({'東向':'E' , '西向':'W' , '南向':'S' , '北向':'N' })
nov = nov.replace({'A3':'3' , 'A2':'2' , 'A1':'1' })
nov.columns = ['ID','Time','Roadname','Direction','Mile','Duration','Severity','Death','Injury']
```

```python
nov = nov[~nov['Time'].isin(rep)]
nov
```
